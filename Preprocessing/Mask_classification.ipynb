```python
now=time.time()
for i,image in enumerate(list_img):
    im = cv.imread('/home/spriyanshu723/biomedical/mask/'+image)
    orignal_img=cv.imread('/home/spriyanshu723/biomedical/size_Output/'+image)
    im=im.astype(float)/255
    orignal_img=orignal_img.astype(float)
    foreground = cv.multiply(im,orignal_img)
    t=np.ones(im.shape)
    background = cv.multiply(1-im,t)
    outImage = cv.add(foreground/255, background)
    plt.imsave('/home/spriyanshu723/biomedical/Mask_classification/'+image,outImage)
    if i+1%20==0:
        print(i+1,'/',len(list_img),'Done')
then=time.time()
print('Time Taken',then-now)
```

## Output Folder Padding using cval=img[0][0][0] 

```python
#!mkdir '/home/spriyanshu723/biomedical/size_Output/'
#!mkdir '/home/spriyanshu723/biomedical/size_GT_Output/'

path = '/home/spriyanshu723/biomedical/Output/'
!ls '/home/spriyanshu723/biomedical/'
```

```python
list_size_h=[]
list_size_w=[]
for images in os.listdir(path):
    img=cv2.imread(path+images)
    list_size_h.append(img.shape[0])
    list_size_w.append(img.shape[1])
```

```python
h=max(list_size_h)
w=max(list_size_w)
h, w
```

```python
import numpy as np
for images in os.listdir(path):
    img=cv2.imread(path+images)
    h_pad=h-img.shape[0]
    w_pad=w-img.shape[1]
    (y1,y2)=(h_pad/2,h_pad/2) if h_pad%2==0 else (h_pad//2,h_pad//2+1)
    (x1,x2)=(w_pad/2,w_pad/2) if w_pad%2==0 else (w_pad//2,w_pad//2+1)
    i = np.pad(img,((int(y1),int(y2)),(int(x1),int(x2)),(0,0)),'constant',constant_values=img[0][0][0])
    plt.imsave('/home/spriyanshu723/biomedical/size_Output/'+images,i)
```

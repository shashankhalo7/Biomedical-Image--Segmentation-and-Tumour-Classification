# Extracting the image for classification using Source and segmentation mask

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
im = cv.imread('/home/witcher/Projects/Biomedical Image- Segmentation and Tumour Classification/BUS/BUS/GT/000022.png')
orignal_img=cv.imread('/home/witcher/Projects/Biomedical Image- Segmentation and Tumour Classification/BUS/BUS/original/000022.png')
im=im.astype(float)/255
orignal_img=orignal_img.astype(float)
```

```python
foreground = cv.multiply(im,orignal_img)
```

```python
plt.imshow(foreground/255)
```

```python
im.shape
```

```python
t=np.ones(im.shape)
```

```python
background = cv.multiply(1-im,t)
```

```python
outImage = cv.add(foreground/255, background)
```

```python
plt.imshow(outImage)
```
